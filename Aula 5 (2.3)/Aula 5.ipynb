# KNN para classificação

KNN (K-Nearest Neighbors) é um algoritmo de classificação de aprendizado supervisionado baseado em instância. Ele é utilizado para prever a classe de uma nova instância, baseando-se nas classes de instâncias vizinhas já conhecidas.

O funcionamento básico do algoritmo é o seguinte:

1. Escolha o número k de vizinhos mais próximos. Esse número é um hiperparâmetro que deve ser escolhido previamente.
2. Calcule a distância entre a nova instância e todas as instâncias do conjunto de dados de treinamento. As distâncias podem ser calculadas usando diferentes métricas, como Euclideana ou Manhattan.
3. Selecione os k vizinhos mais próximos.
4. A classe da nova instância é determinada pela classe mais comum entre os k vizinhos selecionados.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
# adere à maioria
```

```python
import pandas as pd
dados = pd.read_table('/content/drive/MyDrive/Machine Learning - LEAD DELL/Aula 5 (2.3)/fruit_data_with_colors.txt')
dados.info()
```

```python
X = dados[['mass', 'width', 'height', 'color_score']] # dados numericos, atributos
y = dados[['fruit_label']]
```

**separando em treino e teste**

```python
from sklearn.model_selection import train_test_split
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y) # por padrão, 75% para treino e 25% para teste
```

```python
# treino
knn.fit(X_treino, y_treino)
```

```python
# teste
knn.score(X_teste, y_teste)
```

é um bom acerto quando fica acima de 70%

```python
mm = MinMaxScaler()
X_treino = mm.fit_transform(X_treino) # transformando os dados da coluna de treino em 0 ou 1
X_teste = mm.transform(X_teste)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# treinamento
knn.fit(X_treino, y_treino)
```

```python
# teste
knn.score(X_teste, y_teste)
```

# KNN para regressão

KNN (K-Nearest Neighbors) também pode ser usado para regressão, não apenas classificação. A ideia básica é a mesma: dada uma nova instância, o algoritmo encontra os k vizinhos mais próximos no conjunto de treinamento e usa esses vizinhos para prever o valor da nova instância.

A diferença é que, ao invés de prever a classe, o algoritmo está previsão um valor numérico contínuo.

Existem diferentes maneiras de fazer a previsão de valor contínuo, uma das mais comuns é a média dos k vizinhos mais próximos. Outra opção é usar a mediana ou usando um peso baseado na distância dos vizinhos.

Assim como na classificação, a escolha do valor k é importante e pode afetar a precisão da previsão. Além disso, o KNN para regressão é mais sensível a outliers do que o KNN para classificação, é importante tomar cuidado com esses valores fora do padrão.

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=3) #media dos 3 vizinhos mais proximos
```
