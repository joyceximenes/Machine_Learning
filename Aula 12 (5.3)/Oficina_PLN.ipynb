### Bibliotecas necessárias

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

### Pacotes necessários

```python
nltk.download('stopwords')
nltk.download('punkt')
```

Use a variável texto a seguir para identificar as stop words
e após isso aplique o cálculo do **TF-IDF** neste mesmo texto e observe os scores.

**OBS:** Para realizar essa tarefa utilize os módulos estudados na aula

```python
texto = 'Um carro azul seguia rapidamente em uma rodovia, e ao passar por um buraco, o carro furou o pneu, e o motorista desceu do carro azul'
```

**Stopwords**

```python
stw = nltk.corpus.stopwords.words('portuguese')
print(stw)
```

**TF-IDF**

```python
tf_idf = TfidfVectorizer()
matriz = tf_idf.fit_transform([texto]) 
print(matriz)
```

```python
# transformando em vetor
vetor = matriz.todense()
print(vetor)
```

```python
# associar os scores aos nomes das palavras
nomes = tf_idf.get_feature_names()

df = pd.DataFrame(vetor, columns = nomes)
print(df)
```

Assim, os maiores scores se concentraram em 'carro', 'um' e 'azul'.
