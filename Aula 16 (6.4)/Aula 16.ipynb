# PCA (Principal Component Analysis)

Encontrar a aproximação do conjunto de dados original em um espaço de dimensão reduzido

image.png

Objetivos:
* Visualização dos dados
- Redução da complexidade de modelos de Aprendizado de Máquina

Uma das principais informações utilizadas será a variância dos dados <p>
Variancia positiva: quando as duas variaveis crescem uma em relação à outra <p>
Variancia negativa: quando uma decresce em relação à outra
 
> Transformação linear <p>
> Novos atributos são uma combinação linear dos atributos originais

Passos do PCA
1. Toma os dados de entrada
2. Subtração das médias e divisão pelo desvio padrão - Padronização (Standardization)
3. Cálculo da matriz de covariância
4. Cálculo dos autovalores e autovetores (vetores com informações das combinações dos atributos)
5. Seleção dos k maiores autovalores
6. Multiplicação da transposta dos k autovetores pela transposta do conjunto de dados original padronizado (standardized)
7. Novo conjunto de dados

**Com sklearn**

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.datasets import load_boston
```

```python
X,y = load_boston(return_X_y=True)
```

```python
X.shape
```

PCA para dimnuir a quantidade de camadas e aumentar a visualização

```python
pca = PCA(n_components=2) # numero de dimensoes
```

```python
# transformando o X em algo de apenas 2 dimensoes
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1]) 
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca,y)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

perde-se muitas informações, quanto maior a diminuição, menor a acuracia
