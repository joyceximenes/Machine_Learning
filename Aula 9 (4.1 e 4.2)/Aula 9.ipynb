# Overfitting e underfitting

são problemas de generalização. <p>
**Underfitting**: sub-ajuste, ocorre quando o modelo tende a subestimar os resultados. o modelo é simples e o problema tem uma complexidade maior. <p>
quanto mais dados no treinamento, menor o risco de underfitting <p>
image.png

**Overfitting**: sobre ajuste. o modelo de adapta aos dados de treinamento, mas na etapa de testes, com novos dados, ele pode nao generalizar <p>
modelo sofistifacdo para resolver um problema simples <p>

# Tradeoff viés/variancia

"perder de um lado para ganhar de outro" <p>
**Tradeoff** - escolha que pode ter um custo, mas trara um beneficio <p>
**Vies**: quando elevado, o modelo nao ira aprender corretamente(underfitting). quando baixo, os desvios entre observações e predições diminuem. <p>
