```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp = np.array([30, 25, 36, 18, 25, 29, 30, 33, 37, 31, 26, 37, 29, 26, 30, 31, 34, 38])
sorvetes = np.array([20, 12, 50, 10, 18, 25, 26, 32, 48, 22, 16, 52, 24, 20, 28, 29, 35, 40])
dados = pd.DataFrame({'temperatura' : temp, 'numero_sorvetes':sorvetes})
dados.head()
```

```python
# criando o grafico 
plt.plot(dados['temperatura'], dados['numero_sorvetes'], '+')
plt.xlabel('temperatura')
plt.ylabel('numero de sorvetes')
plt.show()
```

```python
# x √© a variavel indepedente e y a dependente
x = dados['temperatura'].to_numpy() # convertendo para float
y = dados['numero_sorvetes'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
# para separar 80% dos dados para treino e 20% para teste

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2)
```

**treinando o modelo**

```python
from sklearn.linear_model import LinearRegression
```

```python
# define um modelo de regressao
modelo = LinearRegression()

# treina o modelo com os dados de treino
modelo.fit(x_treino.reshape(-1, 1), y_treino.reshape(-1, 1)) #reshape coloca os dados na forma de um vetor de n linhas e 1 coluna
```

**realizando a previsao**

```python
# y_previsto cont√©m a previs√£o dos pre√ßos do nosso modelo
y_previsto = modelo.predict(x_teste.reshape(-1,1))
```

```python
plt.plot(range(y_previsto.shape[0]), y_previsto, 'r--')
plt.plot(range(y_teste.shape[0]), y_teste, 'g--')
plt.legend(['Sorvetes previstos', 'Sorvetes vendidos'])
plt.xlabel('√≠ndice')
plt.ylabel('Sorvetes')
plt.show()
```
