```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Machine Learning - LEAD DELL/Aula 7 (2.5)/diabetes.csv')
dados.head()
```

```python
dados.describe()
```

```python
plt.plot(dados['Age'], dados['Pregnancies'], '*')
plt.xlabel('Idade')
plt.ylabel('Gravidez')
plt.show();
```

**separar a variavel dependente**

```python
y = dados['Outcome']
X = dados.drop('Outcome', axis = 1) # tirando da base de dados
```

**separando treino e teste**

```python
from sklearn.model_selection import train_test_split
X_treino, X_teste, y_treino, y_teste = train_test_split(X,y, test_size = 0.2)
```

**selecionando e treinando modelo**

```python
from sklearn.linear_model import LogisticRegression
modelo = LogisticRegression()
```

```python
modelo.fit(X_treino, y_treino)
```

**previsões**

```python
y_previsto = modelo.predict(X_teste)
print(y_previsto[:10])
```

**validando modelo**

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_teste, y_previsto) 
print(accuracy*100)
```
