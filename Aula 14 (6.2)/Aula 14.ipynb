```python
from sklearn.dummy import DummyClassifier, DummyRegressor
from sklearn.datasets import load_boston, load_iris
```

**classifier**

```python
X,y = load_iris(return_X_y=True)
```

```python
dc = DummyClassifier(strategy='stratified') # aleatorio com distribuição igual nos dados de treino
dc.fit(X,y)
dc.score(X,y)
```

**regressor**

```python
X,y = load_boston(return_X_y=True)
```

```python
dr = DummyRegressor(strategy='mean') # media dos rotulos de treinamento
dr.fit(X,y)
dr.score(X,y)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X,y = load_breast_cancer(return_X_y=True)
```

```python
dc = DummyClassifier(strategy='stratified')
dc.fit(X,y)
confusion_matrix(y, dc.predict(X))
```

image.png

```python
print(classification_report(y, dc.predict(X)))
```

**f1-score**: media harmonica entre precisão e recall

```python
from sklearn.dummy import DummyClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
```

```python
X,y = load_iris(return_X_y=True)
dc = DummyClassifier(strategy='stratified')
```

dá o score de cada repartição

```python
cross_val_score(dc, X, y, cv = 3) # cv = qtd de divisões
```

```python
cross_val_score(dc, X, y, cv=3).mean()
```
