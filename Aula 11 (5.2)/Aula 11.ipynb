```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
fashion_mnist = keras.datasets.fashion_mnist

# separando treino e teste das imagens e rotulos
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# forçando os valores a estarem entre 0 e 1
train_images = train_images/256.0
test_images = test_images/256.0
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), #transformando a matriz em um vetor
    keras.layers.Dense(128, activation=tf.nn.relu), # pra ligar o neuronio da camada anterior no da proxima
    # relu: negativo se torna 0, positivo se mantem
    keras.layers.Dense(10, activation=tf.nn.softmax)
    # softmax: define bem as classes
])
model.summary() # quantos pesos ha e a estrutura da rede
```

```python
model.compile(optimizer='adam', # construção dos pesos
              loss='sparse_categorical_crossentropy', # metrica de erro, para trablhar com valores proximos a 0 e 1
              metrics=['accuracy']) # porcentagem de elementos classificados corretamente
```

```python
model.fit(train_images, train_labels, epochs=5) # 5 interações no treinamento
```

```python
# avaliar as imagens e rotulos e guardar as perdas e a acuracia
test_loss, test_acc = model.evaluate(test_images, test_labels)

print('Test accuracy:', test_acc)
```
