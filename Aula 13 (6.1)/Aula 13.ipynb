```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC # svm para classificaÃ§Ã£o
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_treino, X_teste, y_treino, y_teste = train_test_split(X ,y)
```

```python
svm = SVC(kernel='linear', C=1.0) # C Ã© uma medida de quanto o modelo pode errar, mais vai ser ajustada (1.0 Ã© o valor padrao)
```

```python
svm.fit(X_treino, y_treino)
```

```python
svm.score(X_teste, y_teste)
```

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X, y = load_iris(return_X_y=True)
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y)
```

```python
svm = SVC(kernel='linear', C=1.0)
```

```python
svm.fit(X_treino, y_treino)
svm.score(X_teste, y_teste)
```

```python
svm_kernel = SVC(kernel='poly', degree=2) # polinomial, grau 2
svm_kernel.fit(X_treino, y_treino)
svm_kernel.score(X_teste, y_teste)
```

```python
svm_rbf = SVC(kernel='rbf') 
svm_rbf.fit(X_treino, y_treino)
svm_rbf.score(X_teste, y_teste)
```
