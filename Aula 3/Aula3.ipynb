**encontrando outliners** <p>
outliner é um elemento fora do padrão

```python
dados = pd.read_table('/content/drive/MyDrive/Machine Learning - LEAD DELL/Aula 3/fruit_data_with_colors_miss.txt', na_values=['.', '?'])
dados = dados.fillna(dados.mean())
```

```python
laranja = dados[dados['fruit_name'] == 'orange']
est = laranja['mass'].describe()
 
```

```python
# verificando outliners em massa
laranja[(laranja['mass'] > est['mean'] + (est['std']) * 2)] # se a massa é maior que a media da massa + 2 vezes o desvio padrao
# ou seja, estao distantes do padrão
```

```python
laranja[(laranja['mass'] < est['mean'] - (est['std']) * 2)]
# outliners pra baixo
```
