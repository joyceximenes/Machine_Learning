**Implementação**

```python
# importando a rede neural
from sklearn.neural_network import MLPRegressor 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_boston
```

```python
# carregando pras variaveis
X, y = load_boston(return_X_y=True)
# separar treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y)
# formatando
mm = MinMaxScaler()
X_treino = mm.fit_transform(X_treino)
X_teste = mm.transform(X_teste)
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000) # camadas intermediárias; escondidas || max_iter = num maximo de iterações
# 2 camadas, uma com 100 e outra com 50 neuronios 

# treinar a camada
mlp.fit(X_treino, y_treino)
```

```python
mlp.score(X_teste, y_teste)
```
